# cv2.imread()
***

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("287477631.jpg",cv2.IMREAD_GRAYSCALE)
imgb = cv2.imread('287477631.jpg')
```

```python
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.plot([50,100],[80,100],'c',linewidth=5)
plt.show()
```

```python
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.plot([0,100],[0,100],'y',linewidth=5)
plt.show()
```

```python
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.plot([2400,0],[1300,0],'c',linewidth=5)
plt.show()
```

```python
cv2.imshow("111",img)
cv2.imshow('112',imgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('desktop.jpg',img)
```
